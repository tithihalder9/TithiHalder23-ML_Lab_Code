```python
# Linear Regression Example
# Task: Predict CO2 Emission

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# ==================== (a) CO2 Emission Prediction ====================

print("=== (a) CO2 Emission Prediction ===")
fuel_df = pd.read_csv("fuel_consumption_dataset.csv")

# Independent and dependent variables
X_fuel = fuel_df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']]
y_fuel = fuel_df['CO2EMISSIONS']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_fuel, y_fuel, test_size=0.2, random_state=42)

# Train model
model_fuel = LinearRegression()
model_fuel.fit(X_train, y_train)

# Predict
y_pred_fuel = model_fuel.predict(X_test)

# Evaluate
r2_fuel = r2_score(y_test, y_pred_fuel)
mae_fuel = mean_absolute_error(y_test, y_pred_fuel)
mse_fuel = mean_squared_error(y_test, y_pred_fuel)

print(f"R² Score: {r2_fuel:.4f}")
print(f"MAE: {mae_fuel:.4f}")
print(f"MSE: {mse_fuel:.4f}")

# Plot Actual vs Predicted
plt.figure(figsize=(6, 5))
plt.scatter(y_test, y_pred_fuel, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title("CO2 Emission: Actual vs Predicted")
plt.xlabel("Actual CO2 Emission")
plt.ylabel("Predicted CO2 Emission")
plt.grid(True)
plt.show()
```

```python
# Simple Linear Regression Example for a small dataset
# Dataset: housing_price_dataset.csv

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Load dataset
data = pd.read_csv("Area,Bedrooms,Bathrooms,Stories,Par.txt")
data.rename(columns={'Area,Bedrooms,Bathrooms,Stories,Par': 'housing_price_dataset.csv'}, inplace=True)

# Independent and dependent variables
X = data[['Area', 'Bedrooms', 'Bathrooms', 'Stories', 'Parking']]
y = data['Price']

# Train a simple Linear Regression model on the entire dataset
model = LinearRegression()
model.fit(X, y)

# Predict on the entire dataset
y_pred = model.predict(X)

# Evaluate the model
r2 = r2_score(y, y_pred)
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)

print("=== Simple Linear Regression ===")
print(f"R² Score: {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")

# Plot Actual vs Predicted
plt.figure(figsize=(6,5))
plt.scatter(y, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.title("Simple Linear Regression: Actual vs Predicted")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.grid(True)
plt.show()
```
